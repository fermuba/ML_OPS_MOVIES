## Limpieza y extracion de datos de generos de peliculas

```python
# Carga de Librerias de uso
import pandas as pd
import numpy as np
import json
import ast # Abstract Syntax Trees

# configuro semilla de random
np.random.seed(0)
```

```python
# Insgesta del archivos movie_dataset.csv
genero = pd.read_csv('/content/drive/MyDrive/Colab/HENRY/PI-1/Data_font/movies_genres.csv')

# Eliminacion de la columna que se crea con la exportacion de datos
if genero.columns[0] == 'Unnamed: 0':
  genero.drop(columns= 'Unnamed: 0', inplace= True)
```

```python
# Encabezado del Df
genero.head()
```

```python
# Convierto el string en una lista, literal_eval lee la cadena y la convierte en el equivalente python str --> list
genero['genres'] = genero['genres'].apply(ast.literal_eval)
```

```python
# Expandir la columna en múltiples columnas separando los elementos de la lista
columna_expandida = pd.json_normalize(genero['genres'])

# Unir las columnas expandidas al DataFrame original
genero = pd.concat([genero, columna_expandida], axis=1)
```

```python
# Muestra
genero.sample(3)
```

```python
# Informacion
genero.info()
```

```python
# Repetir las expansiones para cada columna nueva
for i in range(0, 8):
  columna_expandida = pd.json_normalize(genero[i])
  genero = pd.concat([genero, columna_expandida], axis= 1)
```

```python
# Nueva informacion
genero.info()
```

```python
# Elimino los id que no necesito
genero.drop(columns= 'id', inplace= True)
for i in range(0 , 8):
  genero.drop(columns= i, axis= 1, inplace= True)
genero.drop(columns='genres', axis= 1, inplace= True)
```

```python
genero.info()
```

```python
# Creaccion de columna con todos los generos disponibles
columnas_generos = genero.iloc[:, 1].unique()
columnas_generos = np.delete(columnas_generos, 11)
columnas_generos.sort()
print(columnas_generos)
```

```python
# Creacion de filtros para cada categoria de genero
action = genero['name'] == 'Action'
adventure = genero['name'] == 'Adventure'
animation = genero['name'] == 'Animation'
comedy = genero['name'] == 'Comedy'
crime = genero['name'] == 'Crime'
documentary = genero['name'] == 'Documentary'
drama = genero['name'] == 'Drama'
family = genero['name'] == 'Family'
fantasy = genero['name'] == 'Fantasy'
foreign = genero['name'] == 'Foreign'
history = genero['name'] == 'History'
horror = genero['name'] == 'Horror'
music = genero['name'] == 'Music'
mystery = genero['name'] == 'Mystery'
romance = genero['name'] == 'Romance'
science_fiction = genero['name'] == 'Science Fiction'
tv_movie = genero['name'] == 'TV Movie'
thriller = genero['name'] == 'Thriller'
war = genero['name'] == 'War'
western = genero['name'] == 'Western'
```

```python
# Creacion de las columnas nuevas para cada categoria de genero
genero['action'] = action.any(axis= 1)
genero['adventure'] = adventure.any(axis= 1)
genero['animation'] = animation.any(axis= 1)
genero['comedy'] = comedy.any(axis= 1)
genero['crime'] = crime.any(axis= 1)
genero['documentary'] = documentary.any(axis= 1)
genero['drama'] = drama.any(axis= 1)
genero['family'] = family.any(axis= 1)
genero['fantasy'] = fantasy.any(axis= 1)
genero['foreign'] = foreign.any(axis= 1)
genero['history'] = history.any(axis= 1)
genero['horror'] = horror.any(axis= 1)
genero['music'] = music.any(axis= 1)
genero['mystery'] = mystery.any(axis= 1)
genero['romance'] = romance.any(axis= 1)
genero['science_fiction'] = science_fiction.any(axis= 1)
genero['tv_movie'] = tv_movie.any(axis= 1)
genero['thriller'] = thriller.any(axis= 1)
genero['war'] = war.any(axis= 1)
genero['western'] = western.any(axis= 1)
```

```python
# Elimina columnas innecesarias
genero.drop(columns= 'name', inplace= True)
genero.head()
```
