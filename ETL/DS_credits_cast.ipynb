## Limpieza y extracion de datos de actores

```python
# Carga de Librerias de uso
import pandas as pd
import numpy as np
import json
import ast # Abstract Syntax Trees

# configuro semilla de random
np.random.seed(0)
```

```python
# Insgesta del archivos movie_dataset.csv
credits_cast = pd.read_csv('/content/drive/MyDrive/Colab/HENRY/PI-1/Data_font/credits_cast.csv')
```

```python
credits_cast.head()
```

```python
# Remplazo nombre de columna id
credits_cast.rename(columns= {'id' : 'id_movie'}, inplace= True)
```

```python
# Convertir la cadena str en una cadena de datos python
credits_cast['cast'] = credits_cast['cast'].apply(ast.literal_eval)
```

```python
# Expando los elementos json uno por columna
columna_expandida = pd.json_normalize(credits_cast['cast'])
credits_cast = pd.concat([credits_cast, columna_expandida], axis= 1)
```

```python
# Impresion del encabezado del Df resultante
credits_cast.head()
```

```python
#Tamaño filas x Columnas
credits_cast.shape
```

```python
# Realizo una nueva expancion a los datos obtenidos
for columna in range(0, 313):
  columna_expandida = pd.json_normalize(credits_cast[columna])
  credits_cast = pd.concat([credits_cast, columna_expandida], axis= 1)
```

```python
# Nuevo tamaño
credits_cast.shape
```

```python
# Impresion del encabezado del Df resultante
credits_cast.head()
```

```python
# Lista de los nombres de las columnas
credits_cast.columns
```

```python
# Eliminacion de columnas que no vamos a usar
columnas_eliminar = np.arange(0 , 313)
credits_cast.drop(columns= columnas_eliminar, inplace= True)
columnas_eliminar = ['Unnamed: 0', 'cast', 'order', 'profile_path',
                     'cast_id', 'character', 'id', 'credit_id',
                     'gender']
credits_cast.drop(columns= columnas_eliminar, inplace= True)
```

```python
# Nuevo tamaño
credits_cast.shape
```

```python
# Creacion de un DataFrame vacio para los nuevos datos, donde solo voy a poner los nombres de todos los actores
credits_cast_new = pd.DataFrame()

# Se obtienen todos los nombres no nulos con su respectivo id_movie de cada columna y se combinan en el nuevo DF
for column in range(1, 314):
  # Obtencion de los indices no nulos
  index = credits_cast.iloc[0: ,column].dropna().index

  # Obtencion de los datos de las columnas no nulas seleccionadas
  id_movie_column = credits_cast['id_movie']
  selected_column = credits_cast.iloc[:, column]

  # Data Frame resultante
  df_result = pd.DataFrame({'id_movie': id_movie_column[index], 'name': selected_column[index]})

  # Todos los datos nuevos se unen en un dataframe de 2 columnas
  credits_cast_new = pd.concat([credits_cast_new , df_result], axis= 0)

```

```python
# Se reinicia el indice del Df nuevo
credits_cast_new.sort_index(inplace= True)
credits_cast_new.reset_index(drop= True)
```

```python
# Filtro por actor
credits_cast_new[credits_cast_new['name'] == 'Tom Hanks']
```

```python
# Eliminacion de duplicados, se comparan por par de datos (id_movie, name)
indices_duplicados = credits_cast_new[credits_cast_new.duplicated(subset= ['id_movie', 'name'], keep= 'first')].index
credits_cast_new.drop(index= indices_duplicados, inplace= True)
```

```python
# Se guardan en archivo los datos obtenidos
credits_cast_new.to_csv('/content/drive/MyDrive/Colab/HENRY/PI-1/Data_font/credits_cast_clean.csv')
```
